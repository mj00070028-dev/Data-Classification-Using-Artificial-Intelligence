```python
# ==============================================================================
# 1. IMPORT LIBRARIES
# ==============================================================================
import os                                    # File system path handling
import warnings                               # Suppress non-critical library warnings

import pandas as pd                             # Data loading and manipulation
import matplotlib.pyplot as plt                  # Plotting
import seaborn as sns                             # Statistical visualizations

from sklearn.model_selection import train_test_split       # Train/test splitting
from sklearn.preprocessing import LabelEncoder               # Target label encoding
from sklearn.tree import DecisionTreeClassifier                # Primary classifier
from sklearn.neighbors import KNeighborsClassifier               # Comparison classifier
from sklearn.linear_model import LogisticRegression                # Comparison classifier
from sklearn.ensemble import RandomForestClassifier                  # Comparison classifier
from sklearn.metrics import (                                          # Evaluation metrics
    accuracy_score,
    confusion_matrix,
    classification_report,
    precision_score,
    recall_score,
    f1_score,
)

warnings.filterwarnings("ignore")  # Suppress sklearn/pandas deprecation noise in console output
```

```python
# ==============================================================================
# GLOBAL CONSTANTS
# ==============================================================================
DATASET_PATH = "Iris.csv"   # Relative path to the dataset file
SCREENSHOTS_DIR = "screenshots"                        # Output directory for visualizations
RANDOM_STATE = 42                                        # Seed for reproducible results
TEST_SIZE = 0.2                                            # 20% of data reserved for testing
FEATURE_COLUMNS = [                                          # Predictive feature columns
    "SepalLengthCm",
    "SepalWidthCm",
    "PetalLengthCm",
    "PetalWidthCm",
]
TARGET_COLUMN = "Species"          # Column containing the class label to predict
SEPARATOR = "=" * 78                 # Reusable console section separator
```

```python
# ==============================================================================
# 2. LOAD DATASET
# ==============================================================================
def load_dataset(dataset_path):
    if not os.path.exists(dataset_path):
        raise FileNotFoundError(
            f"Dataset not found at '{dataset_path}'. "
            f"Ensure 'iris.csv' exists inside the 'dataset/' directory."
        )

    dataframe = pd.read_csv(dataset_path)
    return dataframe
```

```python
# ==============================================================================
# 3. EXPLORE DATASET
# ==============================================================================
def explore_dataset(dataframe):

    print(SEPARATOR)
    print("STEP 1: DATASET EXPLORATION")
    print(SEPARATOR)

    print(f"\nDataset Shape: {dataframe.shape[0]} rows x {dataframe.shape[1]} columns\n")

    print("First 5 Records:")
    print(dataframe.head(), "\n")

    print("Descriptive Statistics (Numeric Features):")
    print(dataframe.describe(), "\n")

    print("Class Distribution (Target Variable):")
    print(dataframe[TARGET_COLUMN].value_counts(), "\n")
```

```python
# ==============================================================================
# 4. DISPLAY DATASET INFORMATION
# ==============================================================================
def display_dataset_information(dataframe):

    print(SEPARATOR)
    print("STEP 2: DATASET STRUCTURE (dtypes & non-null counts)")
    print(SEPARATOR + "\n")
    dataframe.info()
    print()
```

```python
# ==============================================================================
# 5. CHECK MISSING VALUES
# ==============================================================================
def check_missing_values(dataframe):

    print(SEPARATOR)
    print("STEP 3: MISSING VALUE CHECK")
    print(SEPARATOR)

    missing_counts = dataframe.isnull().sum()
    print("\nMissing values per column:")
    print(missing_counts, "\n")

    total_missing = missing_counts.sum()
    if total_missing == 0:
        print("Result: No missing values detected. Dataset is clean.\n")
    else:
        print(f"Result: {total_missing} missing values detected. Consider imputation.\n")

    return missing_counts
```

```python
# ==============================================================================
# 6. SPLIT FEATURES AND TARGET
# ==============================================================================
def split_features_and_target(dataframe):

    features_matrix = dataframe[FEATURE_COLUMNS]

    label_encoder = LabelEncoder()
    encoded_target = label_encoder.fit_transform(dataframe[TARGET_COLUMN])

    print(SEPARATOR)
    print("STEP 4: FEATURE / TARGET SPLIT")
    print(SEPARATOR)
    print(f"\nFeature columns : {FEATURE_COLUMNS}")
    print(f"Target column   : {TARGET_COLUMN}")
    print(f"Encoded classes : {list(label_encoder.classes_)} -> {list(range(len(label_encoder.classes_)))}\n")

    return features_matrix, encoded_target, label_encoder

```

```python
# ==============================================================================
# 7. SPLIT TRAINING AND TESTING DATA
# ==============================================================================
def split_train_test_data(features_matrix, encoded_target):

    X_train, X_test, y_train, y_test = train_test_split(
        features_matrix,
        encoded_target,
        test_size=TEST_SIZE,
        random_state=RANDOM_STATE,
        stratify=encoded_target,
    )

    print(SEPARATOR)
    print("STEP 5: TRAIN / TEST SPLIT")
    print(SEPARATOR)
    print(f"\nTraining samples : {X_train.shape[0]}")
    print(f"Testing samples  : {X_test.shape[0]}")
    print(f"Split ratio      : {int((1 - TEST_SIZE) * 100)}% train / {int(TEST_SIZE * 100)}% test\n")

    return X_train, X_test, y_train, y_test
```

```python
# ==============================================================================
# 8. TRAIN CLASSIFICATION MODELS
# ==============================================================================
def build_candidate_models():

    candidate_models = {
        "Decision Tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
        "Logistic Regression": LogisticRegression(max_iter=200, random_state=RANDOM_STATE),
        "Random Forest": RandomForestClassifier(random_state=RANDOM_STATE),
    }
    return candidate_models


def train_and_evaluate_models(candidate_models, X_train, X_test, y_train, y_test):
    print(SEPARATOR)
    print("STEP 6: MODEL TRAINING & EVALUATION")
    print(SEPARATOR + "\n")

    evaluation_results = {}

    for model_name, model_instance in candidate_models.items():
        # Train the model on the training subset
        model_instance.fit(X_train, y_train)

        # Generate predictions on the unseen test subset
        predicted_labels = model_instance.predict(X_test)

        # Compute standard classification evaluation metrics
        accuracy = accuracy_score(y_test, predicted_labels)
        precision = precision_score(y_test, predicted_labels, average="macro")
        recall = recall_score(y_test, predicted_labels, average="macro")
        f1 = f1_score(y_test, predicted_labels, average="macro")

        evaluation_results[model_name] = {
            "model": model_instance,
            "predictions": predicted_labels,
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "f1_score": f1,
        }

        print(f"[{model_name}]")
        print(f"  Accuracy  : {accuracy:.4f}")
        print(f"  Precision : {precision:.4f}")
        print(f"  Recall    : {recall:.4f}")
        print(f"  F1 Score  : {f1:.4f}\n")

    return evaluation_results


def display_model_comparison_table(evaluation_results):
    print(SEPARATOR)
    print("STEP 7: MODEL COMPARISON SUMMARY")
    print(SEPARATOR + "\n")

    header = f"{'Model':<22}{'Accuracy':<12}{'Precision':<12}{'Recall':<12}{'F1 Score':<12}"
    print(header)
    print("-" * len(header))

    for model_name, results in evaluation_results.items():
        print(
            f"{model_name:<22}"
            f"{results['accuracy']:<12.4f}"
            f"{results['precision']:<12.4f}"
            f"{results['recall']:<12.4f}"
            f"{results['f1_score']:<12.4f}"
        )

    best_model_name = max(evaluation_results, key=lambda name: evaluation_results[name]["accuracy"])
    print(f"\nBest Performing Model: {best_model_name} "
          f"(Accuracy = {evaluation_results[best_model_name]['accuracy']:.4f})\n")

    return best_model_name
```

```python
# ==============================================================================
# 9. DETAILED EVALUATION (CONFUSION MATRIX & CLASSIFICATION REPORT)
# ==============================================================================
def display_detailed_evaluation(best_model_name, evaluation_results, y_test, label_encoder):

    predicted_labels = evaluation_results[best_model_name]["predictions"]
    class_names = label_encoder.classes_

    print(SEPARATOR)
    print(f"STEP 8: DETAILED EVALUATION - {best_model_name}")
    print(SEPARATOR)

    conf_matrix = confusion_matrix(y_test, predicted_labels)
    print("\nConfusion Matrix (rows = actual, columns = predicted):")
    print(pd.DataFrame(conf_matrix, index=class_names, columns=class_names), "\n")

    print("Classification Report:")
    print(classification_report(y_test, predicted_labels, target_names=class_names))

    return conf_matrix
```

```python
# ==============================================================================
# 10. VISUALIZATIONS
# ==============================================================================
def ensure_screenshots_directory():

    os.makedirs(SCREENSHOTS_DIR, exist_ok=True)


def plot_class_distribution(dataframe):

    plt.figure(figsize=(7, 5))
    sns.countplot(data=dataframe, x=TARGET_COLUMN, hue=TARGET_COLUMN,
                   palette="viridis", legend=False)
    plt.title("Class Distribution of Target Species", fontsize=14, fontweight="bold")
    plt.xlabel("Species")
    plt.ylabel("Number of Samples")
    plt.tight_layout()
    plt.savefig(os.path.join(SCREENSHOTS_DIR, "class_distribution.png"), dpi=150)
    plt.close()


def plot_correlation_heatmap(dataframe):
    plt.figure(figsize=(7, 6))
    correlation_matrix = dataframe[FEATURE_COLUMNS].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
    plt.tight_layout()
    plt.savefig(os.path.join(SCREENSHOTS_DIR, "correlation_heatmap.png"), dpi=150)
    plt.close()


def plot_confusion_matrix(conf_matrix, class_names, model_name):

    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f"Confusion Matrix - {model_name}", fontsize=14, fontweight="bold")
    plt.xlabel("Predicted Label")
    plt.ylabel("Actual Label")
    plt.tight_layout()
    plt.savefig(os.path.join(SCREENSHOTS_DIR, "confusion_matrix.png"), dpi=150)
    plt.close()


def plot_feature_importance(model, model_name):

    if not hasattr(model, "feature_importances_"):
        return False

    importance_values = model.feature_importances_
    importance_series = pd.Series(importance_values, index=FEATURE_COLUMNS).sort_values(ascending=True)

    plt.figure(figsize=(7, 5))
    importance_series.plot(kind="barh", color="teal")
    plt.title(f"Feature Importance - {model_name}", fontsize=14, fontweight="bold")
    plt.xlabel("Importance Score")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.savefig(os.path.join(SCREENSHOTS_DIR, "feature_importance.png"), dpi=150)
    plt.close()
    return True


def generate_all_visualizations(dataframe, conf_matrix, label_encoder, best_model_name, evaluation_results):

    print(SEPARATOR)
    print("STEP 9: GENERATING VISUALIZATIONS")
    print(SEPARATOR + "\n")

    ensure_screenshots_directory()

    plot_class_distribution(dataframe)
    print("Saved: screenshots/class_distribution.png")

    plot_correlation_heatmap(dataframe)
    print("Saved: screenshots/correlation_heatmap.png")

    plot_confusion_matrix(conf_matrix, label_encoder.classes_, best_model_name)
    print("Saved: screenshots/confusion_matrix.png")

    # Feature importance is plotted using the Decision Tree model (the
    # project's primary/preferred model) since it reliably supports
    # feature_importances_, regardless of which model wins the comparison.
    decision_tree_model = evaluation_results["Decision Tree"]["model"]
    was_saved = plot_feature_importance(decision_tree_model, "Decision Tree")
    if was_saved:
        print("Saved: screenshots/feature_importance.png\n")
    else:
        print("Feature importance plot skipped (model does not support it).\n")
```

```python
# ==============================================================================
# 11. PREDICTION ON NEW SAMPLE DATA
# ==============================================================================
def predict_new_sample(model, label_encoder, sample_features):

    sample_dataframe = pd.DataFrame([sample_features], columns=FEATURE_COLUMNS)
    encoded_prediction = model.predict(sample_dataframe)[0]
    predicted_species = label_encoder.inverse_transform([encoded_prediction])[0]
    return predicted_species


def demonstrate_new_sample_prediction(best_model_name, evaluation_results, label_encoder):

    print(SEPARATOR)
    print("STEP 10: PREDICTION ON NEW SAMPLE DATA")
    print(SEPARATOR)

    # A new, previously unseen flower measurement (sepal/petal length & width in cm)
    new_sample = [5.9, 3.0, 5.1, 1.8]

    best_model = evaluation_results[best_model_name]["model"]
    predicted_species = predict_new_sample(best_model, label_encoder, new_sample)

    print(f"\nNew Sample Measurements : {dict(zip(FEATURE_COLUMNS, new_sample))}")
    print(f"Model Used              : {best_model_name}")
    print(f"Predicted Species       : {predicted_species}\n")
```

```python
# ==============================================================================
# MAIN PIPELINE ORCHESTRATION
# ==============================================================================
def main():

    print(SEPARATOR)
    print(" DATA CLASSIFICATION USING ARTIFICIAL INTELLIGENCE")
    print(" DecodeLabs Industrial Training Program 2026 - Project 2")
    print(SEPARATOR)

    # Stage 1: Load the dataset
    iris_dataframe = load_dataset(DATASET_PATH)

    # Stage 2: Explore the dataset
    explore_dataset(iris_dataframe)
    display_dataset_information(iris_dataframe)
    check_missing_values(iris_dataframe)

    # Stage 3: Preprocess - split features/target, then train/test
    features_matrix, encoded_target, label_encoder = split_features_and_target(iris_dataframe)
    X_train, X_test, y_train, y_test = split_train_test_data(features_matrix, encoded_target)

    # Stage 4: Train and evaluate candidate models
    candidate_models = build_candidate_models()
    evaluation_results = train_and_evaluate_models(candidate_models, X_train, X_test, y_train, y_test)
    best_model_name = display_model_comparison_table(evaluation_results)

    # Stage 5: Detailed evaluation of the best model
    conf_matrix = display_detailed_evaluation(best_model_name, evaluation_results, y_test, label_encoder)

    # Stage 6: Generate visualizations
    generate_all_visualizations(iris_dataframe, conf_matrix, label_encoder, best_model_name, evaluation_results)

    # Stage 7: Demonstrate prediction on a new, unseen sample
    demonstrate_new_sample_prediction(best_model_name, evaluation_results, label_encoder)

    print(SEPARATOR)
    print(" PIPELINE COMPLETE - All stages executed successfully.")
    print(SEPARATOR)
```

```python
# ==============================================================================
# PROGRAM ENTRY POINT
# ==============================================================================
if __name__ == "__main__":
    main()
```
